```python
import pandas as pd

# Load dataset
df = pd.read_csv("rural_credit_dataset_mixed.csv")

print("Data Loaded Successfully!")
print("Shape:", df.shape)
print(df.head())
```

```python
# Drop ID columns (not useful for ML)
df = df.drop(columns=["borrower_id"])

# Replace missing/NA with 0 for numeric features
df = df.fillna(0)
```

```python
X = df.drop(columns=["loan_repayment_status"])
y = df["loan_repayment_status"]
```

```python

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split

# Example: assume you already have your dataset loaded as X, y
# X = your features DataFrame
# y = your target column (e.g. loan_repayment)

# Define categorical columns
cat_cols = ["gender", "education_level", "village_state", "occupation_type", "irrigation_type", "crop_type"]

# Identify numeric columns automatically
num_cols = [col for col in X.columns if col not in cat_cols]

# --- STEP 1: Detect and fix mixed types in categorical columns ---
for col in cat_cols:
    # Check if column has mixed data types
    types = X[col].apply(type).value_counts()
    if len(types) > 1:
        print(f"Column '{col}' has mixed types {list(types.index)}. Converting all to string.")
    X[col] = X[col].astype(str)

# --- STEP 2: Define column transformer ---
ct = ColumnTransformer([
    ("onehot", OneHotEncoder(drop="first", handle_unknown="ignore"), cat_cols)
], remainder="passthrough")

# --- STEP 3: Apply transformation ---
X_transformed = ct.fit_transform(X)

# --- STEP 4: Train-test split ---
X_train, X_test, y_train, y_test = train_test_split(
    X_transformed, y, test_size=0.2, random_state=42
)

# --- STEP 5: Output shapes ---
print(" Data Preprocessing Complete!")
print("Train Shape:", X_train.shape)
print("Test Shape:", X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler(with_mean=False)  # with_mean=False for sparse data from OneHotEncoder
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#logistic reg
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train, y_train)
y_pred = log_model.predict(X_test)

print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=150, random_state=42)
rf_model.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)

print("Random Forest Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from xgboost import XGBClassifier

xgb_model = XGBClassifier(n_estimators=200, learning_rate=0.1, max_depth=6, random_state=42)
xgb_model.fit(X_train, y_train)
y_pred = xgb_model.predict(X_test)

print("XGBoost Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# Get one-hot encoded column names
ohe_cols = ct.named_transformers_['onehot'].get_feature_names_out(cat_cols)

# Combine with numeric columns
all_feature_names = list(ohe_cols) + num_cols



X_train_df = pd.DataFrame(X_train, columns=all_feature_names)
X_test_df = pd.DataFrame(X_test, columns=all_feature_names)


#ohe_cols = ct.named_transformers_['onehot'].get_feature_names_out(cat_cols)
#shap.dependence_plot("onehot__crop_type_maize", shap_values.values[:, :, 1], X_test_df)

```

```python
import shap
import matplotlib.pyplot as plt

# Use model.predict_proba to avoid XGBoost base_score bug
explainer = shap.Explainer(xgb_model.predict_proba, X_train_df)

# Compute SHAP values
shap_values = explainer(X_test_df)

# Summary plot for class 1
shap.summary_plot(shap_values.values[:, :, 1], X_test_df, feature_names=X_test_df.columns)



```
